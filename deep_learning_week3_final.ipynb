```python
############################################
# Histopathologic Cancer Detection Project #
############################################

# GitHub URL: https://github.com/punampaul04/Deep_Learning

# Description:
#   This notebook demonstrates a complete workflow for the Histopathologic Cancer Detection
#   Kaggle competition. It includes:
#   1. Problem and data description
#   2. Exploratory Data Analysis (EDA)
#   3. Model Architecture and Reasoning
#   4. Training, Tuning, and Results Analysis
#   5. Conclusion and Future Work


############################################
# Import Libraries
############################################

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from tqdm import tqdm
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, confusion_matrix, classification_report
import glob
import io

```

```python
############################################
# Set Paths
############################################

DATA_DIR = "/Users/vedangmehta/Downloads/histopathologic-cancer-detection"
TRAIN_DIR = os.path.join(DATA_DIR, "train")
TEST_DIR = os.path.join(DATA_DIR, "test")
LABELS_PATH = os.path.join(DATA_DIR, "train_labels.csv")

# Ensure directories exist
assert os.path.exists(TRAIN_DIR), f"TRAIN_DIR path does not exist: {TRAIN_DIR}"
assert os.path.exists(TEST_DIR), f"TEST_DIR path does not exist: {TEST_DIR}"
assert os.path.exists(LABELS_PATH), f"LABELS_PATH path does not exist: {LABELS_PATH}"
```

```python
#   1. Problem and data description

############################################
# 1. Problem and Data Description
############################################

df_labels = pd.read_csv(LABELS_PATH)
print("Data Sample:")
display(df_labels.head())

print("\nData Info:")
display(df_labels.info())

print("\nLabel distribution:")
print(df_labels['label'].value_counts())
```

```python
#   2. Exploratory Data Analysis (EDA)

############################################
# 2. Exploratory Data Analysis (EDA)
############################################

plt.figure()
sns.countplot(x='label', data=df_labels)
plt.title('Label Distribution')
plt.show()

# Sample images
sample_ids = df_labels.sample(6, random_state=42)['id'].values

plt.figure(figsize=(10, 6))
for i, img_id in enumerate(sample_ids):
    img_path = os.path.join(TRAIN_DIR, img_id + '.tif')
    img = Image.open(img_path)
    plt.subplot(2, 3, i+1)
    plt.imshow(img)
    plt.title(f"ID: {img_id}")
    plt.axis('off')
plt.suptitle("Sample Images from Training Set")
plt.show()

img_example = Image.open(os.path.join(TRAIN_DIR, df_labels.iloc[0,0]+'.tif'))
print("Example image size and mode:", img_example.size, img_example.mode)

print("Number of images:", len(df_labels))
print("Number of positive samples:", sum(df_labels.label==1))
print("Number of negative samples:", sum(df_labels.label==0))
```

```python
#   3. Model Architecture and Reasoning

############################################
# 3. Data Preparation for Modeling
############################################

# For demonstration, take a smaller subset
sample_frac = 0.1
df_sub = df_labels.sample(frac=sample_frac, random_state=42).reset_index(drop=True)

train_df, val_df = train_test_split(df_sub, test_size=0.1, stratify=df_sub['label'], random_state=42)

print("Training set size:", len(train_df))
print("Validation set size:", len(val_df))

############################################
# 4. Create a TensorFlow Data Pipeline
############################################

BATCH_SIZE = 32
IMAGE_SIZE = (96, 96)

def decode_tif_py_func(image_bytes):
    # image_bytes is a tf.Tensor of type string
    # Convert to numpy bytes and decode with PIL
    image_bytes_val = image_bytes.numpy()
    img = Image.open(io.BytesIO(image_bytes_val))
    return np.array(img)

def load_image(img_id, label):
    img_path = tf.strings.join([TRAIN_DIR, "/", img_id, ".tif"])
    image_bytes = tf.io.read_file(img_path)
    # Use py_function for decoding TIFF with PIL
    image = tf.py_function(decode_tif_py_func, [image_bytes], tf.uint8)
    # Set known shape since we know images are 96x96x3
    image.set_shape([96, 96, 3])
    image = tf.image.convert_image_dtype(image, tf.float32)
    # If needed, resize again if you want to confirm shape
    image = tf.image.resize(image, IMAGE_SIZE)
    return image, label

train_ds = tf.data.Dataset.from_tensor_slices((train_df['id'].values, train_df['label'].values))
val_ds = tf.data.Dataset.from_tensor_slices((val_df['id'].values, val_df['label'].values))

# Map the load_image function
train_ds = train_ds.map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.map(load_image, num_parallel_calls=tf.data.AUTOTUNE)

# Shuffle and batch
train_ds = train_ds.shuffle(1024).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_ds = val_ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

############################################
# 5. Model Architecture
############################################

def build_baseline_cnn():
    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(128, (3,3), activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid')
    ])
    return model

baseline_model = build_baseline_cnn()
baseline_model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss='binary_crossentropy',
    metrics=[tf.keras.metrics.AUC(name='roc_auc')]
)

baseline_model.summary()
```

```python
#   4. Training, Tuning, and Results Analysis

############################################
# 6. Train the Baseline Model
############################################

epochs = 5
history_baseline = baseline_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs
)

plt.figure()
plt.plot(history_baseline.history['roc_auc'], label='train_roc_auc')
plt.plot(history_baseline.history['val_roc_auc'], label='val_roc_auc')
plt.title('Baseline Model ROC_AUC')
plt.xlabel('Epoch')
plt.ylabel('ROC_AUC')
plt.legend()
plt.show()

############################################
# Evaluate on Validation Set
############################################

val_images = val_df['id'].values
val_labels = val_df['label'].values

# Collect predictions
preds = []
for batch_images, batch_labels in val_ds:
    p = baseline_model.predict(batch_images)
    preds.extend(p)

preds = np.array(preds).flatten()
val_roc = roc_auc_score(val_labels, preds)
print(f"Baseline Validation ROC_AUC: {val_roc:.4f}")

threshold = 0.5
binary_preds = (preds >= threshold).astype(int)
cm = confusion_matrix(val_labels, binary_preds)
print("Confusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(val_labels, binary_preds))

############################################
# 7. Transfer Learning (Optional Example)
############################################

from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input

def preprocess_image(image, label):
    image = preprocess_input(image*255.0)
    return image, label

train_ds_tl = train_ds.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
val_ds_tl = val_ds.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)

base_model = EfficientNetB0(
    input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False

model_tl = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])

model_tl.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss='binary_crossentropy',
    metrics=[tf.keras.metrics.AUC(name='roc_auc')]
)

model_tl.summary()

history_tl = model_tl.fit(
    train_ds_tl,
    validation_data=val_ds_tl,
    epochs=epochs
)

plt.figure()
plt.plot(history_tl.history['roc_auc'], label='train_roc_auc')
plt.plot(history_tl.history['val_roc_auc'], label='val_roc_auc')
plt.title('Transfer Learning Model ROC_AUC')
plt.xlabel('Epoch')
plt.ylabel('ROC_AUC')
plt.legend()
plt.show()

preds_tl = []
for batch_images, batch_labels in val_ds_tl:
    p = model_tl.predict(batch_images)
    preds_tl.extend(p)

preds_tl = np.array(preds_tl).flatten()
val_roc_tl = roc_auc_score(val_labels, preds_tl)
print(f"Transfer Learning Validation ROC_AUC: {val_roc_tl:.4f}")

############################################
# 8. Hyperparameter Tuning and Improvements
# (Not shown fully)
############################################
```
